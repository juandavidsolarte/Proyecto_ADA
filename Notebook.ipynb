# Extracción (`extraer_maximo()`)

**Objetivo:** Eliminar y retornar la tarea con mayor prioridad del Max Heap.

**Proceso:**
1. Guardar la raíz.
2. Mover el último elemento a la raíz.
3. Eliminar el último elemento.
4. Restaurar el heap con `heapify_down()`.

**Complejidad:** O(log n)

```python
class MaxHeap:

    def __init__(self):
        self.heap = []

    def mostrar(self):
        for tarea in self.heap:
            print(tarea)


    def insertar(self, tarea):
        """
        Inserta una nueva tarea en el Max Heap.
        """

        # Agregar la tarea al final de la lista
        self.heap.append(tarea)

        # Restaurar la propiedad del Max Heap
        self.heapify_up(len(self.heap) - 1)
    ## ---- HEAPIFY PARA DESPLAZAR LOS NUMEROS
    def heapify_up(self, indice):
        """
        Restaura la propiedad del Max Heap moviendo
        un elemento hacia arriba.
        """

        while indice > 0:

            padre = (indice - 1) // 2

            if self.heap[indice].prioridad > self.heap[padre].prioridad:

                self.heap[indice], self.heap[padre] = (
                    self.heap[padre],
                    self.heap[indice]
                )

                indice = padre

            else:
                break
    def extraer_maximo(self):
        """
        Elimina y retorna la tarea con mayor prioridad.
        """

        # Verificar si el heap está vacío
        if len(self.heap) == 0:
            return None

        # Guardar el elemento de la raíz
        maximo = self.heap[0]

        # Mover el último elemento a la raíz
        self.heap[0] = self.heap[-1]

        # Eliminar el último elemento
        self.heap.pop()

        # Restaurar la propiedad del heap
        if len(self.heap) > 0:
            self.heapify_down(0)

        return maximo
```

# Restauración (`heapify_down()`)

**Objetivo:** Restaurar la propiedad del Max Heap después de una extracción.

**Proceso:**
1. Comparar el nodo con sus hijos.
2. Intercambiar con el hijo de mayor prioridad.
3. Repetir hasta restaurar el heap.

**Complejidad:** O(log n)

```python
class MaxHeap:

    def __init__(self):
        self.heap = []

    def mostrar(self):
        for tarea in self.heap:
            print(tarea)


    def insertar(self, tarea):
        """
        Inserta una nueva tarea en el Max Heap.
        """

        # Agregar la tarea al final de la lista
        self.heap.append(tarea)

        # Restaurar la propiedad del Max Heap
        self.heapify_up(len(self.heap) - 1)
    ## ---- HEAPIFY PARA DESPLAZAR LOS NUMEROS
    def heapify_up(self, indice):
        """
        Restaura la propiedad del Max Heap moviendo
        un elemento hacia arriba.
        """

        while indice > 0:

            padre = (indice - 1) // 2

            if self.heap[indice].prioridad > self.heap[padre].prioridad:

                self.heap[indice], self.heap[padre] = (
                    self.heap[padre],
                    self.heap[indice]
                )

                indice = padre

            else:
                break
    def extraer_maximo(self):
        """
        Elimina y retorna la tarea con mayor prioridad.
        """

        # Verificar si el heap está vacío
        if len(self.heap) == 0:
            return None

        # Guardar el elemento de la raíz
        maximo = self.heap[0]

        # Mover el último elemento a la raíz
        self.heap[0] = self.heap[-1]

        # Eliminar el último elemento
        self.heap.pop()

        # Restaurar la propiedad del heap
        if len(self.heap) > 0:
            self.heapify_down(0)

        return maximo

    def heapify_down(self, indice):
        """
        Restaura la propiedad del Max Heap moviendo
        un elemento hacia abajo.
        """

        while True:

            mayor = indice
            izquierdo = 2 * indice + 1
            derecho = 2 * indice + 2

            # Comparar con el hijo izquierdo
            if (izquierdo < len(self.heap) and
                self.heap[izquierdo].prioridad > self.heap[mayor].prioridad):
                mayor = izquierdo

            # Comparar con el hijo derecho
            if (derecho < len(self.heap) and
                self.heap[derecho].prioridad > self.heap[mayor].prioridad):
                mayor = derecho

            # Si el mayor sigue siendo el padre, terminamos
            if mayor == indice:
                break

            # Intercambiar
            self.heap[indice], self.heap[mayor] = (
                self.heap[mayor],
                self.heap[indice]
            )

            # Continuar desde la nueva posición
            indice = mayor

```

**Complejidad:** O(1)

# Factor de balance (`factor_balance()`)

**Objetivo:** Calcular el equilibrio de un nodo.

**Complejidad:** O(1)


```python
class AVL:

    def __init__(self):

        self.raiz = None

    def obtener_altura(self, nodo):

        if nodo is None:
            return 0

        return nodo.altura

    def actualizar_altura(self, nodo):

        nodo.altura = 1 + max(
            self.obtener_altura(nodo.izquierdo),
            self.obtener_altura(nodo.derecho)
        )
        
    def factor_balance(self, nodo):

        if nodo is None:
            return 0

        return (
            self.obtener_altura(nodo.izquierdo)
            - self.obtener_altura(nodo.derecho)
        )
```

# Rotación Derecha (`rotacion_derecha()`)

**Objetivo:** Reequilibrar el árbol cuando el subárbol izquierdo es más alto.

**Complejidad:** O(1)

# Rotación Izquierda (`rotacion_izquierda()`)

**Complejidad:** O(1)

# Inserción (`insertar()`)

**Objetivo:** Insertar una tarea en el árbol AVL manteniendo el equilibrio.

**Proceso:**
1. Insertar como un árbol binario de búsqueda.
2. Actualizar alturas.
3. Calcular el factor de balance.
4. Aplicar la rotación correspondiente.

**Complejidad:** O(log n)

```python
class AVL:

    def __init__(self):

        self.raiz = None

    def obtener_altura(self, nodo):

        if nodo is None:
            return 0

        return nodo.altura

    def actualizar_altura(self, nodo):

        nodo.altura = 1 + max(
            self.obtener_altura(nodo.izquierdo),
            self.obtener_altura(nodo.derecho)
        )
        
    def factor_balance(self, nodo):

        if nodo is None:
            return 0

        return (
            self.obtener_altura(nodo.izquierdo)
            - self.obtener_altura(nodo.derecho)
        )
    

    def rotacion_derecha(self, y):

        x = y.izquierdo
        T2 = x.derecho

        # Rotación
        x.derecho = y
        y.izquierdo = T2

        # Actualizar alturas
        self.actualizar_altura(y)
        self.actualizar_altura(x)

        return x

    def rotacion_izquierda(self, x):

        y = x.derecho
        T2 = y.izquierdo

        # Rotación
        y.izquierdo = x
        x.derecho = T2

        # Actualizar alturas
        self.actualizar_altura(x)
        self.actualizar_altura(y)

        return y
    
    def insertar(self, tarea):
     self.raiz = self._insertar(self.raiz, tarea)

    def _insertar(self, nodo, tarea):

        # 1. Inserción como un BST
        if nodo is None:
            return NodoAVL(tarea)

        if tarea.id < nodo.id:
            nodo.izquierdo = self._insertar(nodo.izquierdo, tarea)

        elif tarea.id > nodo.id:
            nodo.derecho = self._insertar(nodo.derecho, tarea)

        else:
            return nodo

        # 2. Actualizar altura
        self.actualizar_altura(nodo)

        # 3. Calcular factor de balance
        balance = self.factor_balance(nodo)

        # Caso LL
        if balance > 1 and tarea.id < nodo.izquierdo.id:
            return self.rotacion_derecha(nodo)

        # Caso RR
        if balance < -1 and tarea.id > nodo.derecho.id:
            return self.rotacion_izquierda(nodo)

        # Caso LR
        if balance > 1 and tarea.id > nodo.izquierdo.id:
            nodo.izquierdo = self.rotacion_izquierda(nodo.izquierdo)
            return self.rotacion_derecha(nodo)

        # Caso RL
        if balance < -1 and tarea.id < nodo.derecho.id:
            nodo.derecho = self.rotacion_derecha(nodo.derecho)
            return self.rotacion_izquierda(nodo)

        return nodo
```

```python
class AVL:

    def __init__(self):

        self.raiz = None

    def obtener_altura(self, nodo):

        if nodo is None:
            return 0

        return nodo.altura

    def actualizar_altura(self, nodo):

        nodo.altura = 1 + max(
            self.obtener_altura(nodo.izquierdo),
            self.obtener_altura(nodo.derecho)
        )
        
    def factor_balance(self, nodo):

        if nodo is None:
            return 0

        return (
            self.obtener_altura(nodo.izquierdo)
            - self.obtener_altura(nodo.derecho)
        )
    

    def rotacion_derecha(self, y):

        x = y.izquierdo
        T2 = x.derecho

        # Rotación
        x.derecho = y
        y.izquierdo = T2

        # Actualizar alturas
        self.actualizar_altura(y)
        self.actualizar_altura(x)

        return x

    def rotacion_izquierda(self, x):

        y = x.derecho
        T2 = y.izquierdo

        # Rotación
        y.izquierdo = x
        x.derecho = T2

        # Actualizar alturas
        self.actualizar_altura(x)
        self.actualizar_altura(y)

        return y
    
    def insertar(self, tarea):
     self.raiz = self._insertar(self.raiz, tarea)

    def _insertar(self, nodo, tarea):

        # 1. Inserción como un BST
        if nodo is None:
            return NodoAVL(tarea)

        if tarea.id < nodo.id:
            nodo.izquierdo = self._insertar(nodo.izquierdo, tarea)

        elif tarea.id > nodo.id:
            nodo.derecho = self._insertar(nodo.derecho, tarea)

        else:
            return nodo

        # 2. Actualizar altura
        self.actualizar_altura(nodo)

        # 3. Calcular factor de balance
        balance = self.factor_balance(nodo)

        # Caso LL
        if balance > 1 and tarea.id < nodo.izquierdo.id:
            return self.rotacion_derecha(nodo)

        # Caso RR
        if balance < -1 and tarea.id > nodo.derecho.id:
            return self.rotacion_izquierda(nodo)

        # Caso LR
        if balance > 1 and tarea.id > nodo.izquierdo.id:
            nodo.izquierdo = self.rotacion_izquierda(nodo.izquierdo)
            return self.rotacion_derecha(nodo)

        # Caso RL
        if balance < -1 and tarea.id < nodo.derecho.id:
            nodo.derecho = self.rotacion_derecha(nodo.derecho)
            return self.rotacion_izquierda(nodo)

        return nodo
    def inorden(self):
        self._inorden(self.raiz)
        
    def _inorden(self, nodo):

        if nodo is not None:

            self._inorden(nodo.izquierdo)

            print(nodo.tarea)

            self._inorden(nodo.derecho)

    def eliminar(self, id):
        """
        Elimina una tarea por su identificador.
        """
        self.raiz = self._eliminar(self.raiz, id)    

    def _eliminar(self, nodo, id):

        # No existe
        if nodo is None:
            return nodo

        # Buscar por la izquierda
        if id < nodo.id:
            nodo.izquierdo = self._eliminar(nodo.izquierdo, id)

        # Buscar por la derecha
        elif id > nodo.id:
            nodo.derecho = self._eliminar(nodo.derecho, id)

        # Nodo encontrado
        else:

            # Caso 1: Sin hijo izquierdo
            if nodo.izquierdo is None:
                return nodo.derecho

            # Caso 2: Sin hijo derecho
            elif nodo.derecho is None:
                return nodo.izquierdo

            # Caso 3: El nodo tiene dos hijos
            temporal = self.obtener_minimo(nodo.derecho)

            # Copiar la información del sucesor
            nodo.id = temporal.id
            nodo.tarea = temporal.tarea

            # Eliminar el sucesor
            nodo.derecho = self._eliminar(nodo.derecho, temporal.id)
        
            # Si el nodo fue eliminado
        if nodo is None:
            return nodo

        # Actualizar altura
        self.actualizar_altura(nodo)

        # Calcular balance
        balance = self.factor_balance(nodo)

        # Caso LL
        if balance > 1 and self.factor_balance(nodo.izquierdo) >= 0:
            return self.rotacion_derecha(nodo)

        # Caso LR
        if balance > 1 and self.factor_balance(nodo.izquierdo) < 0:
            nodo.izquierdo = self.rotacion_izquierda(nodo.izquierdo)
            return self.rotacion_derecha(nodo)

        # Caso RR
        if balance < -1 and self.factor_balance(nodo.derecho) <= 0:
            return self.rotacion_izquierda(nodo)

        # Caso RL
        if balance < -1 and self.factor_balance(nodo.derecho) > 0:
            nodo.derecho = self.rotacion_derecha(nodo.derecho)
            return self.rotacion_izquierda(nodo)

        return nodo

        

    def obtener_minimo(self, nodo):
        """
        Retorna el nodo con el menor ID del subárbol.
        """

        while nodo.izquierdo is not None:
            nodo = nodo.izquierdo

        return nodo
```
